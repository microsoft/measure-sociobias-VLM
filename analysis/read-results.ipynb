```python
import pandas as pd
import re
from termcolor import cprint
import copy

df_results = pd.read_excel('../analysis/multimodalbias_all_results.xlsx', sheet_name='results')
df_results_informed = pd.read_excel('../analysis/multimodalbias_all_results.xlsx', sheet_name='resultsinformed')
df_casting = pd.read_excel('../analysis/multimodalbias_all_results.xlsx', sheet_name='resultscasting')
df_casting_informed = pd.read_excel('../analysis/multimodalbias_all_results.xlsx', sheet_name='resultscastinginformed')

prompt_types = {'Uninformed basic':df_results, 'Informed basic':df_results_informed, 'Uninformed indirect (cast director)':df_casting,'Informed indirect (cast director)':df_casting_informed}

model_names = ['llava-1.5-7b-hf', 'bakLlava-v1-hf', 'gemini-pro-vision', 'gpt4v']


def get_model_results(prompt,dataframe):
    cprint(prompt,"red")    
    del dataframe['Unnamed: 0']
    dataframe['fname']=[ele.split(".png")[0].split("a blue humanoid robot is ") [-1]for ele in dataframe['fname']]
    for model in model_names:
        output = dataframe[model]
        out_set = set([])
        for i in range(len(output)):
            ele = output[i]
            if ele=="male" or ele=="female" or ele == "no preference":
                continue
            if re.search("female", ele):#women, feminine
                output[i] = "female"
            elif re.search("male", ele):
                output[i] = "male"
            elif re.search("no preference", ele):
                output[i] = "no preference"
            else:
                out_set.add(ele)
                output[i] = "0"
        #print(out_set)
        dataframe[model] = output
    
    i=0
    for model in model_names:
        i+=1
        distribution_cnt = dataframe[model].value_counts()
        cprint(distribution_cnt*100.0/len(dataframe.index)-1, "blue" if i%2 == 0 else "green")
    del dataframe
```

```python
for prompt in prompt_types.keys():
    get_model_results(prompt=prompt, dataframe=copy.deepcopy(prompt_types[prompt]))
    #print(prompt_types[prompt])
```
